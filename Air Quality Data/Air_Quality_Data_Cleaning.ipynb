## Data Cleaning

```python
data_df.drop(columns=['stn_code','agency','location_monitoring_station'], inplace=True)
data_df.head(2)
```

```python
data_df['rspm'] = data_df['rspm'].replace(np.nan, data_df.rspm.median())
```

```python
data_df['spm'] = data_df['spm'].replace(np.nan, data_df.spm.median())
```

```python
data_df['so2'] = data_df['so2'].replace(np.nan, data_df.so2.median())
```

```python
data_df['no2'] = data_df['no2'].replace(np.nan, data_df.no2 .median())
```

```python
data_df.isna().sum()
```

```python
data_df['pm2_5'] = data_df['pm2_5'].replace(np.nan, ' ')
```

```python
data_df['type'] = data_df['type'].replace(np.nan, ' ')
```

```python
data_df.info(verbose =True)
```

```python
data_df['date'] = pd.to_datetime(data_df['date']).dt.date
```

```python
data_df['year'] = pd.to_datetime(data_df['date']).dt.year
```

```python
data_df['date'] = data_df['date'].replace(np.nan, ' ')
data_df['year'] = data_df['year'].replace(np.nan, ' ')
```

```python
data_df = data_df.dropna(subset=['sampling_date', 'location'])
```
