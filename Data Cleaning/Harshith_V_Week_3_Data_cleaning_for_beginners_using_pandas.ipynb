### **Checking for Missing Values in dataset**

```python
df.isnull().sum()
```

### **There are 7 missing values in Age, have to find the average age in order to fill missing values in age column**

```python
#mean
avg_Age = df['Age'].mean()
avg_Age
```

### **Now, filling the missing values in age column with mean**

```python
df['Age'] = df.Age.round(decimals = 1)
df['Age'] = df.Age.fillna(avg_Age)
df['Age']
```

### **Have to convert float into integer**

```python
df["Age"] = df['Age'].astype(int)
df["Age"]
```

### **And there is only 1 missing value in rating column, handle it by finding avergae**

```python
avg_Rating = df['Rating'].mean()
avg_Rating
```

### **Filling the missing values with avergae**

```python
df['Rating'] = df.Rating.fillna(avg_Rating)
df['Rating'] = df.Rating.round(decimals = 1)
```

### **Now that rating column have negative values have to be replaced to positive**

```python
df['Rating'] = df['Rating'].replace(-1.0, 1.0)
df['Rating']
```

### **Easy Apply Column contains -1 values which can be replaced with FALSE**

```python
df['Easy Apply'] = df['Easy Apply'].replace('-1','FALSE')
df['Easy Apply']
```

### **Replacing $ and represeting in thousands in Salary**

```python
df['Salary'] = df['Salary'].str.replace('$','')
df['Salary'] = df['Salary'].str.replace('k','000')
df['Salary']
```

### **Established column have -1 values need to be replaced with most occurring value in Established column**

```python
df['Established'] = df['Established'].replace(-1,df['Established'].value_counts().index[0])
df['Established']
```

### **Need to clean Location column**

```python
df['Location'] = df['Location'].str.replace(' ',',')
df['Location'] = df['Location'].str.replace('New,','New ')
df['Location']
```

### **Splitting location into Country Name and Country ISO code**

```python
df[['Country Name','Country ISO Code']] = df['Location'].str.split(',', expand = True)
df = df.drop(columns = ['Location'])
df
```

### **Specifying salary as starting salary and ending salary by splitting the Salary column**

```python
df[['Starting Salary','Ending Salary']] = df['Salary'].str.split('-',expand = True)
df = df.drop(columns = ['Salary'])
df
```

### **Have to drop index column as there is default numbering**

```python
df = df.drop(columns = ['Index'])
```
